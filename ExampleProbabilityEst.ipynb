```
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```
dataset=pd.read_csv('/content/drive/My Drive/Sapps_results_2020/NECN-prob-establish-log.csv')
dataset.head()
```

```
grouped=dataset.groupby(['Time',' Species'],as_index=False)[' AvgProbEst'].mean()
#print(grouped)
list1=["QuerPrin","AcerRubr","PinuStro","QuerRubr","QuerAlba","QuerCocc","Lirituli "]
#list1=["LiriTuli" if x=="Lirituli" else x for x in list1]
TopTen=grouped[grouped[' Species'].isin(list1)]

PlotTen=TopTen.pivot(index="Time",
          columns=" Species",
          values=" AvgProbEst")
PlotTen=PlotTen.reindex(columns=list1) 
```

Probability of establishment for major 
species



```
plt.figure(figsize=(20, 10))
x=plt.plot(PlotTen)

plt.legend(PlotTen,loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)

plt.xticks(PlotTen.index);
```
